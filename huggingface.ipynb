```python
import gradio as gr
from transformers import pipeline

# Load GPT-2 model
generator = pipeline("text-generation", model="gpt2")

def generate_text(prompt, length=100, creativity=0.8):
    if not prompt.strip():
        return "Please enter a prompt."

    result = generator(
        prompt,
        max_length=length,
        temperature=creativity,
        num_return_sequences=1
    )

    return result[0]["generated_text"]

# Create simple interface
interface = gr.Interface(
    fn=generate_text,
    inputs=[
        gr.Textbox(
            label="Enter your prompt",
            placeholder="Type something here...",
            lines=3
        ),
        gr.Slider(
            minimum=50,
            maximum=300,
            value=100,
            label="Response length",
            step=10
        ),
        gr.Slider(
            minimum=0.1,
            maximum=1.5,
            value=0.8,
            label="Creativity",
            step=0.1
        )
    ],
    outputs=gr.Textbox(
        label="Generated text",
        lines=8
    ),
    title="GPT-2 Text Generator",
    description="A simple text generator using GPT-2",
    examples=[
        ["Once upon a time"],
        ["The future of AI is"],
        ["In a world where"]
    ]
)

# Launch
interface.launch(share=True)
```
